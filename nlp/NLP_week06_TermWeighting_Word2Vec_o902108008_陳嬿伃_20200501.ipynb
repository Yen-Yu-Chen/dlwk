### Word2Vec in Gensim (2/6), page 38
size:300

upload trained model file to my2, the file name should be recognized.

```python
import logging
from gensim.models import word2vec

logging.basicConfig(format = '%(asctime)s : %(levelname)s :%(message)s', level=logging.INFO)
sentences = word2vec.LineSentence("text8")
model = word2vec.Word2Vec(sentences, size=300)
model.save("word2vec.model")
```

### Word2Vec in Gensim (4/6), page 40 

print out "love", "loving", "loved" 3 words top 10 most similar. 

```python
res = model.most_similar("love", topn = 10)
for item in res:
    print(item[0] + "," + str(item[1]))
```

```python
res = model.most_similar("loving", topn = 10)
for item in res:
    print(item[0] + "," + str(item[1]))
```

```python
res = model.most_similar("loved", topn = 10)
for item in res:
    print(item[0] + "," + str(item[1]))
```

### Word2Vec in Gensim (5/6), page 41
follow to find out those terms' most similar

- King, Queen, Man (this is presented at material)
- Japan, Kyoto, Taiwan
- Sun, Moon, Water 

```python
print("%s AND %s, LIKE %s AND..." %("King", "Queen", "Man"))
res = model.most_similar(["king", "queen"],["man"], topn=10)
for item in res:
    print(item[0] + "," + str(item[1]))
```

```python
print("%s AND %s, LIKE %s AND..." %("Japan", "Kyoto", "Taiwan"))
res = model.most_similar(["japan", "kyoto"],["taiwan"], topn=10)
for item in res:
    print(item[0] + "," + str(item[1]))
```

```python
print("%s AND %s, LIKE %s AND..." %("Sun", "Moon", "Water"))
res = model.most_similar(["sun", "moon"],["water"], topn=10)
for item in res:
    print(item[0] + "," + str(item[1]))
```

### Word2Vec in Gensim (6/6), page 42.

Calculate those terms' cosine similarity,
- king, prince

```python
print("Calculate Cosine Simlarity")
res = model.similarity("king", "prince")
print(res)
```

```python
print("Calculate Cosine Simlarity")
res = model.similarity("male", "female")
print(res)
```
